```python
url="https://www.amazon.in/s?bbn=1389401031&rh=n%3A1389401031%2Cp_36%3A1318505031&dc&qid=1683560871&rnid=1318502031&ref=lp_1389401031_nr_p_36_2"
```

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
response=requests.get(url)
response
```

```python
response.text
```

```python
soup=BeautifulSoup(response.content,'html.parser')
soup.prettify()
```

```python
title=soup.find('title')
title.text
```

```python
amount=soup.find_all('span',class_="a-price-whole")
amount
```

```python
product=soup.find_all("span", class_="a-size-base-plus a-color-base a-text-normal")
product
```

```python
reviews=soup.find_all("span",class_='a-size-base s-underline-text')
```

```python
reviews
```

```python
products=[]
for i in range (0,len(product)):
    products.append(product[i].get_text())
```

```python
products
```

```python
review=[]
for i in range(0,len(reviews)):
    review.append(reviews[i].get_text())
```

```python
review
```

```python
amounts=[]
for i in range(0,len(amount)):
    amounts.append(amount[i].get_text())
```

```python
amounts
```

```python
headings={"Description of a Mobile Phone":products,"AMOUNT":amounts,"No of Reviews":review}
```

```python
df=pd.DataFrame(headings)
```
