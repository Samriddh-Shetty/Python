```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.worldometers.info/gdp/gdp-by-country/"
response=requests.get(url)
response
```

```python
response.text
```

```python
soup=BeautifulSoup(response.content,"html.parser")
print(soup.prettify())
```

```python
title=soup.find("title")
title.text
```

```python
gdp=soup.find("tbody")
gdp
```

```python
body=gdp.find_all("tr")
```

```python
body
```

```python
body_rows=body[0:26]
```

```python
all_rows=[]
for row_num in range(len(body_rows)):
    row = [] 
    for row_item in body_rows[row_num].find_all("td"): 
        pp = row_item.text
        row.append(pp)
    all_rows.append(row[:14])
```

```python
row[:]
```

```python
print(all_rows)
```

```python
for i in all_rows:
    print(i)
```

```python
headings=["Rank","country","GDP(nominal, 2017)","GDP(abbrev.)","GDPgrowth","Population","GDP per capita","Share of World GDP"]
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data=all_rows,columns=headings)
df
```

```python
df.to_csv("gdp.csv",index=False)
```
